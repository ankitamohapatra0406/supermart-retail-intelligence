```python
import pandas as pd
```

```python
df = pd.read_csv("../data/clean_sales.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
category_speed = df.groupby("Category")["Quantity"].sum().sort_values(ascending=False)
category_speed
```

```python
# product movement 
category_speed.to_csv("../outputs/category_movement.csv")
```

```python
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
category_profit
```

```python
# profit analysis
category_profit.to_csv("../outputs/category_profit.csv")
```

```python
# restocking analysis
```

```python
restock = pd.DataFrame({
    "Total_Quantity": category_speed,
    "Total_Profit": category_profit
})
```

```python
restock
```

```python
restock["Recommendation"] = restock["Total_Quantity"].apply(
    lambda x: "Increase Stock" if x > restock["Total_Quantity"].median() else "Reduce Stock"
)
```

```python
restock
```

```python
restock.to_csv("../outputs/restock_recommendations.csv")
```
